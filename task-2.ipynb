```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("/kaggle/input/titanic/train.csv")
df.head()
```

```python
sns.countplot(x="Sex",data=df)
plt.show()
```

```python

sns.catplot(x="Sex", hue="Survived", kind="count", data=df)
sns.catplot(x="Pclass", hue="Survived", kind="count", data=df)
sns.catplot(x="SibSp", hue="Survived", kind="count", data=df)
sns.catplot(x="Parch", hue="Survived", kind="count", data=df)
sns.histplot(x="Age", hue="Survived", data=df, kde=True)

plt.show()
```

```python
sns.histplot(data=df, x='Fare', hue='Survived', kde=True)
plt.show()
```
